# Imports

```python
# Imports
import shap
shap.initjs()
import joblib

# Models
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

## Customization Options
pd.set_option("display.max_columns",100)
plt.style.use(['fivethirtyeight','seaborn-talk'])
mpl.rcParams['figure.facecolor']='white'
```

# Loading Saved Models

```python
loaded = joblib.load('outlet_sales.joblib')
loaded.keys()
```

```python
# Saving each variable
X_train_df = loaded['X_train']
y_train = loaded['y_train']
X_test_df = loaded['X_test']
y_test = loaded['y_test']
preprocessor = loaded['preprocessor']
lin_reg = loaded['LinearRegression']
rf = loaded['RandomForest']
```

# Create X-shap and y-shap Variables

```python
X_shap = shap.sample(X_train_df,nsamples=500,random_state=321)
X_shap.head()
```

```python
# get the corresponding y-values
y_shap = y_train.loc[X_shap.index]
y_shap.head()
```

# Random Forest Model

```python
explainer = shap.Explainer(rf)
explainer
```

```python
# Getting the shap values
shap_values = explainer(X_shap,y_shap)
print(type(shap_values))
shap_values.shape
```

## Visualizing the Summary Plot

```python
# visualizing the summary plot with bar
fig, ax = plt.subplots(figsize=(8,8))
shap.summary_plot(shap_values,features= X_shap,plot_type='bar',show=False)
ax.set_title('RandomForestClassifier');
fig.savefig('tree_shap_bar.png', dpi=100, bbox_inches='tight')
```



- Outlet Establishment Year is different in the SHAP summary plot vs. the the top 5 most important features. The top 2 are also the same, but supermarket type 3 is third most important in the SHAP plot, whereas item visibility is third in the important features. Item weight is not in the top 5 in the SHAP plot.

```python
# visualizing the summary plot with dot
fig, ax = plt.subplots(figsize=(8,8))
shap.summary_plot(shap_values,features= X_shap,plot_type='dot',show=False)
ax.set_title('RandomForestClassifier');
fig.savefig('tree_shap_dot.png', dpi=100, bbox_inches='tight')
```

- For the Item_MRP, many of the samples positively affected outlet sales, as seen by the many red dots to the right of 0, although it does appear balanced on the left as well with the blue dots. Grocery store negatively impacted the outlet sales, as seen by the many red dots to the left. Supermarket type 3 had a positive impact on the outlet sales.
